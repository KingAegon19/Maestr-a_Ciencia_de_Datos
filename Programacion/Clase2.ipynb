El Rover Curiosity en Marte nos ha enviado una cadena de telemetría, pero la transmisión es ruidosa. Tu tarea es escribir una función en Python llamada analizar_telemetria(datos_str, umbral_voltaje, umbral_temp) que procese esta cadena y reporte las anomalías.

La cadena que recibirás es:

V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;
Tu función debe hacer lo siguiente:

Ignorar cualquier paquete de datos que esté mal formado o cuyos valores numéricos sean inválidos (ej. 'XYZ').
Ignorar valores de sensores que marquen un error obvio (voltajes negativos como -999 o temperaturas extremas como 999).
Calcular el voltaje promedio y la temperatura máxima de los paquetes de datos válidos.
Contar cuántos paquetes válidos tienen un voltaje por debajo de 4.58V (este valor te será pasado en el argumento umbral_voltaje).
Contar cuántos paquetes válidos superan una temperatura de 21.60°C (este valor te será pasado en el argumento umbral_temp).
La función debe devolver un diccionario con las siguientes llaves: 'voltaje_promedio', 'temp_maxima', 'alertas_voltaje', y 'alertas_temp'. Si no hay datos válidos para calcular un promedio o un máximo, debe devolver 0 para ese valor.

```python
def analizar_telemetria(datos_str, umbral_voltaje, umbral_temp):
    paquetes = datos_str.strip().rstrip(';').split(';')
    print(paquetes)
    voltajes = []
    temperaturas = []
    alertas_voltaje = 0
    alertas_temp = 0
    
    for paquete in paquetes:
        print(paquete)
        try:
            campos = paquete.split(',')
            if len(campos) != 3:
                continue
            
            v = float(campos[0].split('=')[1])
            t = float(campos[1].split('=')[1])
            
            # Ignorar valores con errores obvios
            if v < 0 or t > 100:
                continue
            
            voltajes.append(v)
            temperaturas.append(t)
            
            if v < umbral_voltaje:
                alertas_voltaje += 1
            if t > umbral_temp:
                alertas_temp += 1
                
        except (ValueError, IndexError):
            continue
    
    voltaje_promedio = sum(voltajes) / len(voltajes) if voltajes else 0
    temp_maxima = max(temperaturas) if temperaturas else 0
    
    return {
        'voltaje_promedio': voltaje_promedio,
        'temp_maxima': temp_maxima,
        'alertas_voltaje': alertas_voltaje,
        'alertas_temp': alertas_temp
    }
    
    
# Ejemplo de uso
datos = "V=5.24,T=21.20,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=3.24,T=20.75,C=914;V=2.24,T=27.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;V=5.24,T=20.75,C=914;"

resultado = analizar_telemetria(datos, umbral_voltaje=5.0, umbral_temp=25.0)
print(resultado)
```
